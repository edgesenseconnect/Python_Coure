```python
from tkinter import *
class MyWindow:
    def __init__(self, win):
        self.lbl1=Label(win, text='First Name')
        self.lbl2=Label(win, text='Last Name')
        self.lbl3=Label(win, text='Result')
        self.t1=Entry(bd=3)
        self.t2=Entry()
        self.t3=Entry()
        self.btn1 = Button(win, text='Welcome')
        self.btn2=Button(win, text='Subtract')
        self.lbl1.place(x=100, y=50)
        self.t1.place(x=200, y=50)
        self.lbl2.place(x=100, y=100)
        self.t2.place(x=200, y=100)
        self.b1=Button(win, text='Welcome', command=self.add)
        self.b2=Button(win, text='Subtract',command=self.sub)
        #self.b2.bind('<Button-1>', self.sub)
        self.b1.place(x=100, y=150)
        self.b2.place(x=200, y=150)
        self.lbl3.place(x=100, y=200)
        self.t3.place(x=200, y=200)
    def add(self):
        num1=(self.t1.get())
        num2=(self.t2.get())
        result=num1+num2
        self.t3.insert(END, str(result +"  "+"welcome"+" "+str(datetime.datetime.now())))
    def sub(self):
        num1=int(self.t1.get())
        num2=int(self.t2.get())
        result=num1-num2
        self.t3.insert(END, str(result))

window=Tk()
mywin=MyWindow(window)
window.title('Hello Python')
window.geometry("400x300+10+10")
window.mainloop()
```

```python
#!/usr/bin/python3
import datetime
import tkinter as tk
import pyqrcode
import json
import png
import os


from PIL import Image, ImageTk


fields = ('AssetType', 'AssetID', 'AssetDept', 'AssetOwner', 'ShelfLife')

def jsondata(entries):
    # period rate:
    AssetType = entries['AssetType'].get()
    
    # principal loan:
    AssetDept = entries['AssetDept'].get()
    AssetID =  entries['AssetID'].get()
    ShelfLife = entries['ShelfLife'].get()
    json_data = {'AssetDept':entries['AssetDept'].get(),
                'AssetType':entries['AssetType'].get(),
                'AssetID': entries['AssetID'].get(),
                'ShelfLife':entries['ShelfLife'].get(),
                'EntryDate':str(datetime.datetime.now())}
    #print(json_data)
    global preview_screen
    global root
    preview_screen = tk.Toplevel(root)
    preview_screen.title("Preview Data")
    preview_screen.geometry("300x250")
    tk.Label(preview_screen, text=json_data,wraplength = 300).pack()
    

def qrcode(entries):
    global root
    json_data = {'AssetDept':entries['AssetDept'].get(),
                'AssetType':entries['AssetType'].get(),
                'AssetID': entries['AssetID'].get(),
                'ShelfLife':entries['ShelfLife'].get(),
                'EntryDate':str(datetime.datetime.now())}
     #print(json_data)
    url = pyqrcode.create(json.dumps(json_data))
    url.png(entries['AssetID'].get()+".png", scale=8)
    image = Image.open(entries['AssetID'].get()+".png")
    photo = ImageTk.PhotoImage(image)
    
    global qr_screen
    global root
    qr_screen = tk.Toplevel(root)
    qr_screen.title("QR Code")
    qr_screen.geometry("500x250")
    label = tk.Label(qr_screen).pack()
    label = tk.Label(qr_screen,image=photo)
    label.image = photo # keep a reference!
    label.pack()
    

def makeform(root, fields):
    entries = {}
    for field in fields:
        #print(field)
        row = tk.Frame(root)
        lab = tk.Label(row, width=22, text=field+": ", anchor='w')
        ent = tk.Entry(row)
        ent.insert(0, "0")
        row.pack(side=tk.TOP, 
                 fill=tk.X, 
                 padx=5, 
                 pady=5)
        lab.pack(side=tk.LEFT)
        ent.pack(side=tk.RIGHT, 
                 expand=tk.YES, 
                 fill=tk.X)
        entries[field] = ent
    #print("sdfsdfsdf",entries)
    return entries

if __name__ == '__main__':
    global root
    root = tk.Tk()
    ents = makeform(root, fields)
    b1 = tk.Button(root, text='QR Code',
           command=(lambda e=ents:qrcode(ents)))
    b1.pack(side=tk.LEFT, padx=5, pady=5)
    b2 = tk.Button(root, text='Preview',
           command=(lambda e=ents: jsondata(e)))
    b2.pack(side=tk.LEFT, padx=5, pady=5)
    b3 = tk.Button(root, text='Quit', command=root.quit)
    b3.pack(side=tk.LEFT, padx=5, pady=5)
    root.mainloop()
```

```python
#!/usr/bin/python3
import datetime
import tkinter as tk
import pyqrcode
import json
import png
import os
import boto3
#from PIL import Image, ImageTk
import os

###########################################AWS S3 bucket########################
ACCESS_KEY =  'xxxxxxxxxxxxx'
SECRET_KEY = 'xxxxxxxxxxxxxxxxx'
S3_LOCATION = 'http://{}.s3.amazonaws.com/'.format('iotadaka')
s3 = boto3.client(
"s3",
aws_access_key_id=ACCESS_KEY,
aws_secret_access_key=SECRET_KEY
)
bucket_resource = s3
bucket_name = "iotadaka"
###########################################AWS S3 bucket########################

fields = ('AssetType', 'AssetID', 'AssetDept', 'AssetOwner', 'ShelfLife')

def jsondata(entries):
    # period rate:
    AssetType = entries['AssetType'].get()
    
    # principal loan:
    AssetDept = entries['AssetDept'].get()
    AssetID =  entries['AssetID'].get()
    ShelfLife = entries['ShelfLife'].get()
    json_data = {'AssetDept':entries['AssetDept'].get(),
                'AssetType':entries['AssetType'].get(),
                'AssetID': entries['AssetID'].get(),
                'ShelfLife':entries['ShelfLife'].get(),
                'EntryDate':str(datetime.datetime.now()),
                "AssetOwner":entries['AssetOwner'].get()}
    #print(json_data)
    global preview_screen
    global root
    preview_screen = tk.Toplevel(root)
    preview_screen.title("Preview Data")
    preview_screen.geometry("300x250")
    tk.Label(preview_screen, text=json_data,wraplength = 300).pack()
    

def qrcode(entries):
    from PIL import Image
    from PIL import ImageFont
    from PIL import ImageDraw 
    global root
    global bucket_resource
    global bucket_name 
    json_data = {'AssetDept':entries['AssetDept'].get(),
                'AssetType':entries['AssetType'].get(),
                'AssetID': entries['AssetID'].get(),
                'ShelfLife':entries['ShelfLife'].get(),
                'EntryDate':str(datetime.datetime.now()),
                "AssetOwner":entries['AssetOwner'].get()}
     #print(json_data)
    url = pyqrcode.create(json.dumps(json_data))
    filename = entries['AssetID'].get()+".png"
    url.png(filename, scale=4,quiet_zone=10)
    image = Image.open(filename)#entries['AssetID'].get()+".jpg")
    draw = ImageDraw.Draw(image)
    #font = ImageFont.truetype(<font-file>, <font-size>)
    font = ImageFont.truetype("abel-regular.ttf", 16)
    #draw.text((x, y),"Sample Text",(r,g,b))
    draw.text((0, 0),entries['AssetID'].get()+":"+entries['AssetType'].get()+'\n'+json_data['EntryDate'],(0),font=font)
    image.save(filename)
    
    #############################uplad to S3###########################
    bucket_resource.upload_file(Bucket = bucket_name,
                                Filename=filename,
                                Key=filename
                                )
    ###################################################################
    photo = ImageTk.PhotoImage(image)
    
    global qr_screen
    global root
    qr_screen = tk.Toplevel(root)
    qr_screen.title("QR Code")
    qr_screen.geometry("500x250")
    label = tk.Label(qr_screen).pack()
    label = tk.Label(qr_screen,image=photo)
    label.image = photo # keep a reference!
    label.pack()
    

def makeform(root, fields):
    entries = {}
    for field in fields:
        #print(field)
        row = tk.Frame(root)
        lab = tk.Label(row, width=22, text=field+": ", anchor='w')
        ent = tk.Entry(row)
        ent.insert(0, "0")
        row.pack(side=tk.TOP, 
                 fill=tk.X, 
                 padx=5, 
                 pady=5)
        lab.pack(side=tk.LEFT)
        ent.pack(side=tk.RIGHT, 
                 expand=tk.YES, 
                 fill=tk.X)
        entries[field] = ent
    #print("sdfsdfsdf",entries)
    return entries

if __name__ == '__main__':
    global root
    root = tk.Tk()
    ents = makeform(root, fields)
    b1 = tk.Button(root, text='QR Code',
           command=(lambda e=ents:qrcode(e)))
    b1.pack(side=tk.LEFT, padx=5, pady=5)
    b2 = tk.Button(root, text='Preview',
           command=(lambda e=ents: jsondata(e)))
    b2.pack(side=tk.LEFT, padx=5, pady=5)
    b3 = tk.Button(root, text='Quit', command=root.quit)
    b3.pack(side=tk.LEFT, padx=5, pady=5)
    root.mainloop()
```
